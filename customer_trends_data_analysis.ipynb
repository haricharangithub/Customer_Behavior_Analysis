```python
import pandas as pd
```

```python
dataset = pd.read_csv("customer_shopping_behavior.csv")
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset["Review Rating"] = dataset.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
dataset.isnull().sum()
```

```python
dataset.columns
```

```python
dataset.columns = dataset.columns.str.lower()
dataset.columns = dataset.columns.str.replace(" ","_")
```

```python
dataset.columns
```

```python
dataset = dataset.rename(columns = {"purchase_amount_(usd)":"purchase_amount"})
```

```python
dataset.columns
```

```python
# creating a column age_group
groups = ["Young Adult","Adult","Middle_age","Senior"]
dataset["age_group"] = pd.qcut(dataset["age"],q=4,labels = groups)
```

```python
dataset[["age","age_group"]].head(9)
```

```python
dataset["frequency_of_purchases"].unique()
```

```python
# creating a column purchase_frequencey_days

frequency_matching = {
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
    
}
dataset["purchesing_frequency_days"] = dataset["frequency_of_purchases"].map(frequency_matching)
```

```python
dataset[["frequency_of_purchases","purchesing_frequency_days"]].head(10)
```

```python
dataset[["discount_applied","promo_code_used"]].head(10)
```

```python
dataset = dataset.drop("promo_code_used",axis = 1)
```
